```
import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.preprocessing import MultiLabelBinarizer
```

```
bgg_data = pd.read_csv('data/data_sources_bronze/bgg_dataset_20260527.csv')
```

```
# Function to split a single column into multiple binary columns

def binarize(df: pd.DataFrame, column: str):
    """
    Expecting specific column to contain strings of comma separated values.
    Returns a pd.DataFrame of given column split and binarized.
    """
    
    split_column = df[column].fillna('').str.split(', ')

    mlb = MultiLabelBinarizer()

    binary_array = mlb.fit_transform(split_column)

    binary_df = pd.DataFrame(
        binary_array,
        columns=mlb.classes_,
        index=df['id']
    )

    return binary_df.reset_index()
```

```
# binarize the columns we are interested in
# drop the first unnamed column, a result of filling nas

game_mechanics_binary = binarize(bgg_data, 'mechanics').drop(columns='')
game_categories_binary = binarize(bgg_data, 'categories').drop(columns='')
```
